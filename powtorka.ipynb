## Podstawowe Struktury danych używane w PANDAS
### Series
* Opis: Jednowymiarowa struktura danych podobna do listy lub tablicy. Można myśleć o niej jak o kolumnie w arkuszu kalkulacyjnym, z wartościami i indeksami.

* Indeksowanie: Każdy element w Series ma przypisany unikalny indeks, który może być liczbą lub tekstem.

* Typ danych: Wszystkie wartości w Series muszą być tego samego typu.

### DataFrame
* Opis: Dwuwymiarowa struktura danych, przypominająca tabelę z wierszami i kolumnami (jak arkusz kalkulacyjny). Każda kolumna jest obiektem Series.

* Indeksowanie: Wiersze i kolumny mają swoje indeksy, a dane można selekcjonować na podstawie indeksów wierszy i nazw kolumn.

* Różnorodne typy danych: W przeciwieństwie do Series, różne kolumny w DataFrame mogą zawierać różne typy danych (np. jedna kolumna może przechowywać liczby całkowite, a inna tekst).

## Zaawansowane metody na kolumnach

```python
df[['month']].rank()  # nie wiem czy kiedys bede tego uzywać
```

```python
df[['name']].shift(2) # przsuwa wartości o liczbe którą podamy w funkcji. Obliczanie zmian miedzy wierszami
```

```python
df[['arr_time']].cumsum() #dodawana jest wartosc bieżąca oraz wszystkie poprzednie
```

```python
df['arr_time'].rolling(window=5).mean().head(30) # srednia z 5 wierszy które przemieszczaja sie o jeden i zapisuja kolejna wartosc z kolejnych 5 wierszy
```

```python
df[['flight']].clip(1000, 2000).head(10) # ogranicza wartosci w wierszach od 1000 do 2000
#zamienia wartosci wieksze niz 2000 na 2000 a mniejsze niz 1000 na 1000
```

```python
df[['flight']].head(10)
```

```python
df.groupby('name')[['dep_time']].mean() # bardzo ciekawe narzędzie!!!!
# tutaj zgrupowałam wszystkei dep_time po nazwach lini lotniczych i obliczyłam dla nich srednia
```
